# **Popularity Based Recommender System**

```python
books.head()
```

```python
ratings.head()
```

```python
users.head()
```

```python
#joining books and user ratings into one table
books_with_name=ratings.merge(books,on='ISBN')
```

```python
books_with_name.head()
```

```python
books_with_name.shape
```

```python
popular_df=books_with_name.groupby('Book-Title').agg(num_of_ratings=('Book-Rating','count'),avg_rating=('Book-Rating','mean'))
```

```python
popular_df=popular_df.reset_index()
```

```python
popular_df
```

```python
popular_df.sort_values('num_of_ratings',ascending=False)
```

```python
#Popularity is based on the on of people read the book ("num_rating">300)
#it is based on the rating it got
popular_df=popular_df[popular_df['num_of_ratings']>300].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df
```

```python
#for the model deployment I need Book-title,Author,Image URL
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_of_ratings','avg_rating']]
```

**Colabrative Filtering**
## -similar book prediction based on users feedback

```python
books_with_name.head()
```

```python
#Grouping based on user-id tells the no of books rated by each user:
x=books_with_name.groupby('User-ID').count()
x
```

```python
x.index
```

```python
x.shape
```

```python
#Select only user who atleast gave feed back for 200 books(Power users)
x=x[x['Book-Rating']>=200].index
x
```

```python
x.shape
```

```python
x.sort_values()
```

```python
#Selecting only records of power users
filtered_ratings=books_with_name[books_with_name["User-ID"].isin(x)]
filtered_ratings
```

```python
# I am cosidering only the best users(atleast 200 books feedback) group them on the book title.
y=filtered_ratings.groupby('Book-Title').count()
y
```

```python
#the above dataframe tell how many users have read the book.
y.sort_values("User-ID")
```

```python
y=y["User-ID"]>=50
y
```

```python
famous_books=y[y].index
famous_books
```

```python
final_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
final_ratings
```

```python
# pivot table giving the ratings for each book from each user
#Book row with userid as column
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
pt
```

```python
similarity_scores=cosine_similarity(pt)
similarity_scores
```

```python
df_temp=pd.DataFrame(similarity_scores)
df_temp
```

```python
pt.index
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    # let create empty list and in that list i want to populate with the book information
    # Book author book-title image url
    # empty list
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
recommend('1984')
```

```python
recommend('Harry Potter and the Chamber of Secrets (Book 2)')
```
